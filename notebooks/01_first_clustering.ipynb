# First Clustering

```python
# import libraries
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# create dataset with make_blobs
X, y = make_blobs(n_samples=100, centers=4, cluster_std=[1, 1.5, 2, 2], random_state=7)

# Convert to DataFrame for easier handling and visualization
df_blobs = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df_blobs['Cluster'] = y

# Quick visualization
plt.scatter(df_blobs['Feature_1'], df_blobs['Feature_2'], c=df_blobs['Cluster'], cmap='viridis', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Blobs Dataset')
plt.show()
```

# Clustering Model

```python
from sklearn.cluster import KMeans
```

```python
# define model

model_kmeans = KMeans(n_clusters=4, random_state=7)
```

```python
y_pred = model_kmeans.fit_predict(X)
```

```python
# plot the results
fig, axs = plt.subplots(1, 2, figsize=(20, 12))

# Original clusters
axs[0].scatter(df_blobs['Feature_1'], df_blobs['Feature_2'], c=df_blobs['Cluster'], cmap='viridis', s=50)
axs[0].set_title('Original Clusters')

# KMeans clusters
axs[1].scatter(df_blobs['Feature_1'], df_blobs['Feature_2'], c=y_pred, cmap='viridis', s=50)
axs[1].set_title('KMeans Clusters')

plt.show()
```
